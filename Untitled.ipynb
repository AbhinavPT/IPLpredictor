```python
match_df1 = match_df[['match_id' , 'city','winner','total_runs']]
```

```python
delivery_df = match_df1.merge(deliveries , on ='match_id')
```

```python
delivery_df['batting_team'] = delivery_df['batting_team'].str.replace('Delhi Daredevils', 'Delhi Capitals')
delivery_df['bowling_team'] = delivery_df['bowling_team'].str.replace('Delhi Daredevils', 'Delhi Capitals')
delivery_df['winner'] = delivery_df['winner'].str.replace('Delhi Daredevils', 'Delhi Capitals')

delivery_df['winner'] = delivery_df['winner'].str.replace('Deccan Chargers' ,'Sunrisers Hyderabad')
delivery_df['batting_team'] = delivery_df['batting_team'].str.replace('Deccan Chargers' ,'Sunrisers Hyderabad')

delivery_df['bowling_team'] = delivery_df['bowling_team'].str.replace('Deccan Chargers' ,'Sunrisers Hyderabad')
```

```python
delivery_df = delivery_df[delivery_df['inning']==2]
delivery_df
```

```python
delivery_df['bowling_team'].unique()
```

```python
delivery_df['balls_left'] = 120 - 6*(delivery_df['over']-1)-delivery_df['ball']
```

```python
delivery_df.loc[:, 'current_score'] = delivery_df.groupby(by=['match_id'])['total_runs_y'].cumsum()
delivery_df
```

```python
delivery_df['total_runs_y'].unique()
```

```python
delivery_df.info()
```

```python
delivery_df['player_dismissed'] = np.where(delivery_df['player_dismissed'].isna(), 'Not Dismissed', delivery_df['player_dismissed'])
delivery_df['player_dismissed'] = np.where(delivery_df['player_dismissed'] == 'Not Dismissed', 0, 1).astype('int64')
delivery_df['wickets_left'] = delivery_df.groupby(by = ['match_id'])['player_dismissed'].cumsum()
delivery_df['wickets_left'] = 10 - delivery_df['wickets_left'] 
```

```python
delivery_df
```

```python
delivery_df['crr'] = 6*delivery_df['current_score']/(121 -delivery_df['balls_left'])
```

```python
delivery_df['rrr'] = 6*(delivery_df['total_runs_x'] +1-delivery_df['current_score'])/(delivery_df['balls_left'] +1)
```

```python
delivery_df.tail(123)
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result , axis=1)
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']+1
delivery_df
```

```python
delivery_df = delivery_df[delivery_df['balls_left'] != 0]
delivery_df = delivery_df[delivery_df['runs_left'] >= 0]
delivery_df
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','total_runs_x','balls_left','runs_left','wickets_left','crr','rrr','result']]
```

```python
final_df
```

```python
final_df = final_df.sample(final_df.shape[0])
final_df
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace = True)
final_df.shape
```

```python
final_df = final_df[final_df['balls_left'] != 0]
```

```python
final_df.shape
```

```python
final_df.describe()
```

```python
def update_ballsleft(rrr):
    if rrr < 0:
        return 0
    else:
        return rrr

final_df['balls_left'] = final_df['balls_left'].apply(update_ballsleft)
```

```python
def update_rrr(rrr):
    if rrr > 36:
        return 50
    elif rrr < 0:
        return 0
    else:
        return rrr

final_df['rrr'] = final_df['rrr'].apply(update_rrr)
final_df.describe()
```

```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y , test_size =0.2, random_state =1)
```

```python
X_train
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

trf = ColumnTransformer([
    ('trf' , OneHotEncoder(sparse_output = False,drop ='first'),['batting_team','bowling_team','city'])
],remainder = 'passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
X_train.describe()
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
